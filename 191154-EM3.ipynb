```python
# People per Birthmonth sorted by calendar numeric month
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
cleanrecweek['birthmonth'] = pd.Categorical(cleanrecweek['birthmonth'], categories=months, ordered=True)
cleanrecweek['birthmonth'].sort_values(ascending=True)
cleanrecweek["birthmonth"].value_counts().reset_index(name='counts').rename(columns={'index':'birthmonth'}).sort_values(by='birthmonth',ascending=True).reset_index(drop=True)
```

```python
# write code here
def carrier (phonenumber):
    if phonenumber[0:4]=="0918":
        return "Smart"
    else:
        return "Globe"
```

```python
cleanrecweek["carrier"]=cleanrecweek["phonenumber"].apply(carrier)
cleanrecweek[["gender","carrier"]]
```
